### HTTP

**H**yper **T**ext **T**ransfer **P**rotocol
- HTTP is a communications protocol that provides a structure for requests between the client and the server on a network.
- For example, the web browser on the user's computer (the client) uses the HTTP protocol to request information from a website on a server.

We will use APIs that employ the **HTTP** protocol as transport.

### URL
Contains information about the resource being requested from the SERVER.
Examples:

- https://www.google.com/webhp?authuser=2
- https://www.towardsdatascience.com
- https://www.ironhack.com/
    - Protocol: https (https == http is the same, but https is encrypted)
    - Domain Name --> IRONHACK
    - TLD (Top-Level Domain) --> .com

## Web Petitions, APIs

### Requests

The **Requests** are questions that we pose to a SERVER to receive information (a **response**).
Types of requests (verb):
 * GET: read the information of the resource, without modifying it in any way. Accessing the website from the browser **gets** information.
 * POST
 * PUT
 * DELETE
 * PATCH
 * ...

VERB (what to do) : GET, POST, PUT, DELETE

HEADERS (metadata) : User-Agent, Accept-Language

REST (Representational State Transfer) : based on HTTP protocol, RESTful APIs



### Response

The responses can be of various types:
 * a JSON (similar to a Python dictionary)
 * an image
 * a video
 * an HTML page
 * a PDF file
 * ...

Responses follow the following structure. They have:
 * a HEADER section: contains metadata about the server's behavior
 * a BODY: contains the desired information

An important part of the HEADER is the **status code**. This code is a numerical code that indicates the server's result. There are different status codes depending on whether the server has managed to carry out the request or has not managed to do anything. These are some groups of status codes:

- **2xx successful**: the request was successfully received, understood, and accepted
- **3xx redirection**: more actions are required to complete the request
- **4xx client error**: the request contains incorrect syntax or cannot be fulfilled
- **5xx server error**: the server has failed to complete an apparently valid request

Complete list:
https://en.wikipedia.org/wiki/List_of_HTTP_status_codes

## API

### API

**A**pplication **P**rogramming **I**nterface
* En la era de la nube, los programas necesitan comunicarse a través de Internet.
* Dentro de un proyecto, el backend puede querer compartir información con el frontend sin acceso a la base de datos.
* Es un contrato entre productos de software. 
* Es un manual de instrucciones para utilizar un software.
* El cliente llama al servidor a través de una API, el servidor responde.

### RESTful APIs

La mayoría de las API interactivas siguen la lógica de REST, una serie de reglas   
- [Principios rectores de REST](https://restfulapi.net/)   
- [Qué es una API RESTful (vídeo)](https://www.youtube.com/watch?v=JD6VNRdGl98&ab_channel=LeonardoKuffo)   

Quién hace qué:  
* Nosotros (cliente) enviamos una petición
* Ellos (servidor) devuelven una respuesta, generalmente en formato JSON      

Las APIs siempre tienen que ofrecer documentación de sus diferentes servicios: **endpoints**  
Cada endpoint es una URL diferente. 
A veces tenemos que pasar **parametros** a un endpoint de una API. Parecido a cuando le pasamos parámetros a una función de Python. 

La respuesta más típica de una API es en formato JSON. La diferencia entre un **JSON** y el diccionario al que estamos acostumbrados es que el JSON solo almacena información en formato *string*.

En este punto del Bootcamp es cuando **leer documentación se vuelve esencial**, y es que cada API  "funcionará" de una forma diferente y para poder usarla necesitaremos saber lo que le necesita.

### Headers

HTTP Headers son una parte importante de una API request y response, y representan la meta-data asociada con la  API request y response. Llevan información de:

* Request and Response Body
* Request Authorization
* Response Caching 
* Response Cookies

**EL TOKEN**

- Para algunas APIs necesitamos un *token*, es una identificará como la persona que esta haciendo las llamadas a la API. 
    

⚠️🚨 ¡Cuidado con el token! ⚠️🚨  El token es una credencial personal para acceder a la API, a través de la cual se gestiona tu cuota de solicitudes a ésta. Por ello, lo ideal es seguir un procedimiento de seguridad y guardar el token como variable en un documento .env y almacenar el nombre de este archivo en un .gitignore. [NewsAPI documentation](https://newsapi.org/docs/authentication) menciona respecto a la autenticación:

La autenticación se realiza mediante una API key.

La puedes incluir en tu request de una de las siguientes maneras:

* apiKey como parametro de la query
* Mediante el parámetro X-Api-Key en el HTTP header.
* Por autorización de la HTTP header.

Se recomienda una de las últimas 2 opciones para que la API key no sea visible a otros en logs o "request sniffing".

```python
def get_pokemon(id_number):
    res=requests.get('https://pokeapi.co/api/v2/pokemon/{}'.format(id_number))
    #res = requests.get(f"https://pokeapi.co/api/v2/pokemon/{id_number}")
    
    if(res.status_code==200):
        data = res.json()
        return {"id": id_number,'name':data['name'], "peso":data["weight"],'image':data['sprites']['front_default']}
    else:
        raise ValueError("Cannot get pokemon, id does not exist")
```

## Glosario

* DNS: domain name system.
* HTTP: is the protocol used to transfer data over the web.
* API: application programming interface.
* REST: series of rules, architectural style.

* `requests`: Python library for interacting with APIs.
* URL: server name you want to ask information for.
* endpoint: server service you want to ask information for.
* parameters: extra parameters to your query.
* headers: metadata, invisible information.
